```python
import pandas as pd
```

```python
df = pd.read_excel('IMDB Movie Ratings.xlsx')
```

```python
stats = df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Year'], df['Rating'])
```

```python
plt.title('Scatter Plot of Year vs. Rating')
plt.xlabel('Year')
plt.ylabel('Rating')

plt.show()
```

```python
plt.plot(df.groupby('Year')['Rating'].mean())
```

```python
plt.title('Average Rating Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
df['decade']=(df['Year']//10)*10
plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
plt.title('Distribution of Ratings by Decade')
plt.xlabel('Decade')
plt.ylabel('Rating')
plt.boxplot([df[df['decade']==decade]['Rating'] for decade in df['decade'].unique()],labels=df['decade'].unique())
plt.show()
```

```python
correlation = df['Year'].corr(df['Rating'])
print(f'Correlation between year and rating: {correlation: .2f}')
```
